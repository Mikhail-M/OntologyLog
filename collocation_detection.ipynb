считываем вопросы mail.ru из кулинарных категорий

```python
question = pd.read_csv('merged.csv', usecols=['question', 'category_urlname']).dropna()
```

```python
food_categories = ['sweets_baking', 'drinks', 'food_devices', 'main_courses',
       'food_other', 'celebration', 'food_howto', 'food_buy', 'salads',
       'food_preserving', 'soups', 'food_kid']
questions = question.question
food_questions = question[np.in1d(question.category_urlname, food_categories)].question
```

Вычленим биграммы с высоким порогом

```python
class Generator(object):
    def __init__(self, data):
        self.data = data
        self.tokenizer = WordPunctTokenizer()
        self.lem = Mystem()

    def __iter__(self):
        for row in self.data.values:
           yield self.tokenizer.tokenize(''.join(self.lem.lemmatize(row.lower())))
corpus = Generator(food_questions)
```

```python
from gensim.models.phrases import Phraser
```

```python
%%time
bigram = gensim.models.Phrases(corpus, min_count=2, threshold=100, delimiter=b' ')
bigram_phraser = Phraser(bigram)
#tokens__ = trigram_phraser[tokens_]
#all_words = [i for j in tokens__ for i in j]
#all_words
```

```python
%time trigram = gensim.models.Phrases(bigram_phraser[corpus], min_count=3, threshold=100, delimiter=b' ')
```

```python
trigram_phraser = Phraser(trigram)
```

```python
%time fourgram = gensim.models.Phrases(trigram_phraser[bigram_phraser[corpus]], min_count=3, threshold=100, delimiter=b' ')
```

```python
%time all_fourgram = [p[0].decode("utf-8", "strict") for p in fourgram.export_phrases(trigram_phraser[bigram_phraser[corpus]])]    
```
